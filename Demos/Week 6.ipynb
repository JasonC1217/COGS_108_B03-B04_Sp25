## Part I: Data Wrangling

Here, we have a predefined function to standarize heights data:

````python
def standardize_height(str_in):
    """
    Convert a height string to centimeters.
    
    This function takes a height string as input, which can be in inches or centimeters,
    and returns the height standardized to centimeters. If the input is invalid or not
    recognized, it returns a NaN value.
    
    Parameters:
    - str_in (str): A string representing a height, e.g., "5 inches" or "12 cms".
    
    Returns:
    - float: The height in centimeters or NaN if input is invalid.
    """
    
    try:
        # Convert input string to lowercase to ensure consistent processing
        str_in = str_in.lower()
        
        # Remove any leading or trailing whitespaces
        str_in = str_in.strip()

        # Check if input string contains 'in' or 'inches' indicating height in inches
        if 'in' in str_in or 'inches' in str_in:
            # Remove the word 'inches' and 'in' from the string to isolate the number
            str_in = str_in.replace('inches', '')
            str_in = str_in.replace('in', '')
            
            # Convert the isolated number string to float and then convert inches to centimeters
            output = float(str_in)
            output = 2.54 * output
        # Check if input string contains 'cms' or 'centimeters' indicating height in centimeters
        elif 'cms' in str_in or 'centimeters' in str_in:
            # Remove the word 'cms' and 'centimeters' from the string to isolate the number
            str_in = str_in.replace('cms', '')
            str_in = str_in.replace('centimeters', '')
            
            # Convert the isolated number string to float
            output = float(str_in)
        else:
            # If neither 'inches' nor 'cms' is present, return NaN
            output = np.nan
    except:
        # If any error occurs during processing (e.g., converting a non-number string to float), return NaN
        output = np.nan
        
    return output
````

Now we have a function defined. We can just apply this function to all elements in a single column using .apply():

````python
df['height'] = df['height'].apply(standardize_height)
````

Now we have all the heights data standardized, what about weight?

### Specifically, we will need:
#### Input
A string representing a weight, e.g.:
- `"150 pounds"`
- `"68 kgs"`

#### Output
A `float` representing the weight in kilograms, or `np.nan` if the input is invalid.

#### Logic
1. The input string is converted to lowercase and stripped of extra spaces.
2. If the string contains:
   - `"pounds"` or `"lbs"` → the numeric part is converted from pounds to kilograms using `1 lb = 0.453592 kg`.
   - `"kilograms"` or `"kgs"` → the numeric part is used directly as kilograms.
3. If none of the known units are found or conversion fails, the function returns `np.nan`.

#### Edge Case Handling

Uses a `try-except` block to catch formatting errors or invalid numbers.
Returns `np.nan` for unrecognized formats or conversion failures.

#### Examples
```python
standardize_weight("150 pounds")    # ➝ 68.0388
standardize_weight("70 kgs")        # ➝ 70.0
standardize_weight("invalid input") # ➝ nan


````python
def standardize_weight(str_in):
    """
    Convert a weight string to kilograms.
    
    This function takes a weight string as input, which can be in pounds or kilograms,
    and returns the weight standardized to kilograms. If the input is invalid or not
    recognized, it returns a NaN value.
    
    Parameters:
    - str_in (str): A string representing a weight, e.g., "150 pounds" or "68 kgs".
    
    Returns:
    - float: The weight in kilograms or NaN if the input is invalid.
    """
    
    try:
        # Convert the input string to lowercase to ensure consistent processing
        str_in = str_in.lower()
        
        # Remove any leading or trailing whitespaces
        str_in = str_in.strip()

        # Check if the input string contains 'pounds' or 'lbs' indicating weight in pounds
        if 'pounds' in str_in or 'lbs' in str_in:
            # Remove the word 'pounds' and 'lbs' from the string to isolate the number
            str_in = str_in.replace('pounds', '')
            str_in = str_in.replace('lbs', '')
            
            # Convert the isolated number string to float and then convert pounds to kilograms
            output = float(str_in)
            output = 0.453592 * output
            
        # Check if the input string contains 'kgs' or 'kilograms' indicating weight in kilograms
        elif 'kgs' in str_in or 'kilograms' in str_in:
            # Remove the word 'kgs' and 'kilograms' from the string to isolate the number
            str_in = str_in.replace('kgs', '')
            str_in = str_in.replace('kilograms', '')
            
            # Convert the isolated number string to float
            output = float(str_in)
        else:
            # If neither 'pounds' nor 'kgs' is present, return NaN
            output = np.nan
    except:
        # If any error occurs during processing (e.g., converting a non-number string to float), return NaN
        output = np.nan
        
    return output
````

````python
df['weight'] = df['weight'].apply(standardize_weight)
````

#### What is a t-test?

A **t-test** is a statistical hypothesis test used to determine whether there is a **significant difference between the means** of two groups. It is especially useful when dealing with **small sample sizes** and when the population standard deviation is **unknown**. The test assumes that the data is approximately **normally distributed**.

There are several types of t-tests, including:
- **Independent t-test (unpaired):** Compares means from two **independent** groups (e.g., control vs. treatment).
- **Paired t-test:** Compares means from the **same group at different times** (e.g., before vs. after treatment).
- **One-sample t-test:** Compares the mean of a **single group** to a known or hypothesized population mean.

#### T-Test Formula

For an **independent two-sample t-test**, the test statistic is calculated as:

#### $t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

Where:
- $ \bar{X}_1, \bar{X}_2 $: sample means of group 1 and group 2  
- $ s_1^2, s_2^2 $: sample variances of group 1 and group 2  
- $ n_1, n_2 $: sample sizes of group 1 and group 2  

This formula compares the difference in means relative to the **pooled standard error**.

#### Interpretation
- The **null hypothesis (H₀)** typically states that there is **no difference** between the group means:  
  $ H_0: \mu_1 = \mu_2
  $
- The **alternative hypothesis (H₁)** suggests that the means **are different**:  
  $
  H_1: \mu_1 \ne \mu_2
  $

A **large absolute value of the t-statistic** (along with a small p-value) indicates strong evidence **against** the null hypothesis.

#### When to Use a T-Test
- Comparing test scores of two classes
- Evaluating the effect of a treatment
- Testing marketing campaign outcomes across user groups

> Note: Always check assumptions (normality, equal variance, independence) before applying a t-test.

#### OLS: Ordinary Least Squares Regression

While height and weight may contribute to athletic potential, a player's actual **performance** is best captured by their `overall` score. In this analysis, we will explore whether **nationality**—specifically being from **France** versus **Brazil**—is associated with differences in players’ `overall` ratings.

We'll use **Ordinary Least Squares (OLS)** regression from the `statsmodels` library to evaluate this relationship.

#### Hypotheses

We set up the following hypotheses:

- **Null Hypothesis ($H_0$):**  
  There is **no relationship** between nationality and overall score.  
  $
  H_0: \beta_{\text{France}} = \beta_{\text{Brazil}}
  $

- **Alternative Hypothesis ($H_1$):**  
  Nationality **does** affect overall score.  
  $
  H_1: \beta_{\text{France}} \ne \beta_{\text{Brazil}}
  $

#### Ordinary Least Squares (OLS): Concept and Objective

**Ordinary Least Squares (OLS)** is a method used in linear regression to find the best-fitting line (or hyperplane) through a set of data points. It works by **minimizing the Mean Squared Error (MSE)** — the average of the squared differences between observed and predicted values.

#### Objective of OLS

The goal of OLS is to find the coefficients $ \boldsymbol{\beta} $ that minimize the **residual sum of squares**, or equivalently, the **mean squared error**.

Given:
- $ Y $: vector of observed values (dependent variable)
- $ \mathbf{X} $: matrix of input features (including a column of 1s for the intercept)
- $ \boldsymbol{\beta} $: vector of coefficients to estimate
- $ \hat{Y} = \mathbf{X} \boldsymbol{\beta} $: predicted values
- $ \varepsilon = Y - \hat{Y} $: residuals (errors)

#### Optimization Problem

OLS solves the following minimization problem:

$
\min_{\boldsymbol{\beta}} \ \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \mathbf{x}_i^\top \boldsymbol{\beta} \right)^2
$

Or, in compact matrix form:

$
\min_{\boldsymbol{\beta}} \ \frac{1}{n} \| \mathbf{Y} - \mathbf{X} \boldsymbol{\beta} \|^2
$

This minimizes the **mean squared error** between predicted and observed values.

![OLS_example](ols.png) 

#### Closed-form Solution

The optimal OLS estimate of $ \boldsymbol{\beta} $ is given by:

$
\hat{\boldsymbol{\beta}} = \left( \mathbf{X}^\top \mathbf{X} \right)^{-1} \mathbf{X}^\top \mathbf{Y}
$

#### Why It Matters

- OLS provides the **Best Linear Unbiased Estimator**.
- It is widely used for both prediction and inference in statistics and machine learning.
- The model is interpretable, fast to compute, and foundational for more complex techniques.
