From the dataset above, we can see that the numbers in daily circulation are all in string type and are separated with commas. In order to perform numerical analysis, we will need to remove the commas and convert the strings to floats.

`.str.replace(',', '')` can be used to remove all commas from the strings in this column. This is necessary because numeric values formatted as strings often use commas as thousands separators, which need to be eliminated for numerical operations.

````python
#get rid of the commas in the numbers for Daily Circulation

pulitzer['Daily Circulation, 2004'] = pulitzer['Daily Circulation, 2004'].str.replace(',', '').astype(float)

pulitzer['Daily Circulation, 2013'] = pulitzer['Daily Circulation, 2013'].str.replace(',', '').astype(float)
pulitzer['Change in Daily Circulation, 2004-2013'] = pulitzer['Change in Daily Circulation, 2004-2013'].str.replace('%', '').astype(float)
````

What you should see here is that there appears to be some general positive relationship but note that there are outliers in circulation, where circulation is much higher than number of Pulitzers won as well as outliers in Pulitzers won. Overall, most of the data are clustered around zero Pulitzers. This is because the distributions of both of our variables are skewed right. This suggests that we may need to transform these data to use linear regression to ensure that the large outlier values are not driving our relationship.

As discussed in class, we're applying a `log10` transformation to the Pulitzer Prize count data. This transformation helps normalize the data and reduce the influence of extremely large values.

We'll use an **offset of 0.1** to avoid taking the logarithm of zero (which is undefined) and to prevent extreme compression of low values.

#### Why use an offset?

The offset helps us avoid two issues:

1. **Log(0) is undefined**: If any Pulitzer counts are exactly zero, applying `log10` directly would cause a math error.
2. **Better spread of small values**: Using a small offset like 0.1 (rather than 1) prevents low values from collapsing too close to zero on the log scale, preserving some variation.

````python
# log transform with a 0.1 offset

pulitzer['pulitzer_log10'] = np.log10(pulitzer['Pulitzer Prize Winners and Finalists, 1990-2014'] + 0.1)
````

````python
# visualize the distribtuion of the log10 column
with plt.xkcd():
    sns.histplot(pulitzer['pulitzer_log10'], bins=30, color="dimgrey")
    ax1.set_xlabel('1990-2003')
    ax1.set_ylabel('Number of Pulitzers');
    plt.show()
````

Let's observe the relationship again:

````python
with plt.xkcd():
    sns.lmplot(x = 'pulitzer_log10',
               y = 'Change in Daily Circulation, 2004-2013',
               data = pulitzer, 
               fit_reg = True, 
               height = 6,
               aspect = 2, 
               line_kws={'color': 'red'},
               scatter_kws={'color': 'red'})
    plt.xlabel('Pulitzer winners and finalists (1990-2014)', size=18)
    plt.ylabel('Change in print and digital circulation (2004-2013)', size=18)
    plt.title('');
    plt.show()
````

And let's perform a linear regression:

In this analysis, we’re examining whether the number of Pulitzer Prizes a newspaper has won (on a log10 scale) is associated with changes in its **daily circulation** between 2004 and 2013.

We will:

1. **Subsets relevant columns** from the `pulitzer` DataFrame:
   - `'Change in Daily Circulation, 2004-2013'`: the response variable.
   - `'pulitzer_log10'`: the predictor variable, log-transformed with an offset.
   
2. **Renames the columns** to simplify model syntax:
   - `circulation`: change in readership.
   - `pulitzer_log10`: log-scaled Pulitzer count.

3. **Fits a linear regression model** using `statsmodels.formula.api.ols`:
   ```python
   res_1 = smf.ols("circulation ~ pulitzer_log10", data=df).fit()
   ```

````python
df = pulitzer[['Change in Daily Circulation, 2004-2013', 
                         'pulitzer_log10']]
df.columns = ['circulation', 'pulitzer_log10']
df.head()

res_1 = smf.ols("circulation ~ pulitzer_log10", data=df).fit()

# Check out the results
print(res_1.summary())
````

### What did you see?

#### Statistical Significance:
- **p-value for `pulitzer_log10`**: 0.101  
  This is **not statistically significant** at the 0.05 level, meaning we *cannot confidently conclude* that Pulitzer count predicts circulation change in this sample.
  
- **R-squared**: 0.055  
  Only about 5.5% of the variance in circulation change is explained by Pulitzer count. The model has limited explanatory power.

#### Final Notes:
- While the trend is in the expected direction (more Pulitzers = less circulation loss), the result is not statistically significant.
- Consider adding more variables (e.g. ownership, digital presence) or examining subgroups to better explain the variation in circulation change.
